###  Gráfico de Barras Colorido

Situação: Uma loja deseja visualizar o desempenho de vendas mensal.
Dados: meses = ['Jan', 'Fev', 'Mar', 'Abr'], vendas = [200, 150, 300, 220].
Biblioteca: Matplotlib
Métodos: Use plt.bar(meses, vendas, color=['red', 'green', 'blue', 'yellow']) para criar barras coloridas.

```python
import matplotlib.pyplot as plt
meses = ['Jan', 'Fev', 'Mar', 'Abr']
vendas = [200,150,300,220]
plt.bar(meses,vendas,color=['red','green','blue','yellow'])
```

### Gráfico de Pizza com Legenda

Situação: Um supermercado quer entender as preferências de frutas dos clientes.
Dados: frutas = ['Maçã', 'Banana', 'Laranja', 'Manga'], quantidade = [10, 15, 5, 20].
Biblioteca: Matplotlib
Métodos: Utilize plt.pie(quantidade, labels=frutas, autopct='%1.1f%%') e plt.legend() para o gráfico e legenda.


```python
import matplotlib.pyplot as plt
frutas = ['Maçã', 'Banana', 'Laranja', 'Manga']
quantidade = [10, 15, 5, 20]

plt.pie(quantidade,labels=frutas, autopct='%1.1f%%')
plt.legend()
```

### Gráfico de Barras Empilhadas

Situação: Comparação de vendas de produtos entre regiões.
Dados: produtos = ['Produto A', 'Produto B'], norte = [30, 25], sul = [40, 35].
Biblioteca: Matplotlib
Métodos: Use plt.bar(produtos, norte) e plt.bar(produtos, sul, bottom=norte) para empilhar as vendas.

```python
import matplotlib.pyplot as plt
produtos = ['Produto A', 'Produto B']
norte = [30, 25]
sul = [40, 35]
plt.bar(produtos,norte)
plt.bar(produtos,sul, bottom=norte)
```

### Personalizando Cores em Gráficos de Barras

Situação: Um departamento de marketing quer destacar informações em um gráfico.
Dados: categorias = ['Cat 1', 'Cat 2', 'Cat 3'], valores = [100, 200, 150].
Biblioteca: Matplotlib
Métodos: plt.bar(categorias, valores, color=['#aabbcc', '#ffaacc', '#ccaaff']) e plt.gcf().set_facecolor('#dddddd') para cores personalizadas.

```python
categorias = ['Cat1','Cat2','Cat3']
valores = [100,200,150]
plt.bar(categorias,valores,color=['#aabbcc','#ffaacc','#ccaaff'])
plt.gcf().set_facecolor('#dddddd')
```

### Gráfico com Múltiplos Eixos

Situação: Análise financeira de vendas e lucro.
Dados: meses = ['Jan', 'Fev', 'Mar', 'Abr'], vendas = [100, 150, 200, 250], lucro = [20, 30, 50, 40].
Biblioteca: Matplotlib
Métodos: Crie o gráfico de barras com plt.bar(meses, vendas) e adicione um eixo Y secundário para o lucro com plt.twinx() e plt.plot(meses, lucro).


```python
meses = ['Jan', 'Fev', 'Mar', 'Abr']
vendas = [100, 150, 200, 250]
lucro = [20, 30, 50, 40]
plt.bar(meses,vendas)
plt.twinx()
plt.plot(meses,lucro)
```

### BeautifulSoup - Web Scraping

Situação: Extrair dados de um site de notícias para análise de tendências.
Biblioteca: BeautifulSoup
Métodos: (requestBeautifulSoups.get(url).text, 'html.parser'), .find(), .find_all()
Resolução: Use BeautifulSoup para analisar o HTML e extrair informações relevantes como títulos de notícias ou links.

```python
from bs4 import BeautifulSoup as bs
import requests
resposta = requests.get('https://github.com/NandoLFC12')
site = bs(resposta.text, 'html.parser')
links = site.find_all('a')
for link in links:
  print(link.get('href'))
```

### Pillow - Processamento de Imagens

Situação: Redimensionar e modificar a cor de uma imagem para uma galeria online.
Biblioteca: Pillow (PIL)
Métodos: Image.open(), .resize(), .convert()
Resolução: Abra uma imagem, redimensione-a e converta-a para escala de cinza.


```python
from PIL import Image

# Abrir a imagem
imagem = Image.open('4693ae38-a001-467e-8e07-e2c4428f639e.jpg')  # verifique a extensão

# Redimensionar
imagem = imagem.resize((800, 600))

# Converter para RGB
final = imagem.convert('RGB')

# Salvar
final.save('imagem_final.jpg')
```
